```python
import pandas as pd

# Replace with your GitHub raw link
url = "https://raw.githubusercontent.com/Kishore3148/Smart-Grid-Monitoring-System/refs/heads/main/Smart_Grid_Monitoring_System.csv"

# Load CSV data
df = pd.read_csv(url)

# Convert Timestamp to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Display first few rows
df.head()

#Line Chart - Voltage & Frequency Over Time
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df["Timestamp"], df["Voltage (V)"], marker='o', linestyle='-', color='b', label="Voltage (V)")
plt.plot(df["Timestamp"], df["Frequency (Hz)"], marker='s', linestyle='--', color='r', label="Frequency (Hz)")
plt.xticks(rotation=45)
plt.xlabel("Time")
plt.ylabel("Voltage & Frequency")
plt.title("Voltage & Frequency Over Time")
plt.legend()
plt.grid()
plt.show()

#Bar Chart - Renewable Energy Contribution Over Time
plt.figure(figsize=(12, 6))
plt.bar(df["Timestamp"], df["Renewable Energy (%)"], color='green', label="Renewable Energy (%)")
plt.xticks(rotation=45)
plt.xlabel("Time")
plt.ylabel("Renewable Energy (%)")
plt.title("Renewable Energy Contribution Over Time")
plt.legend()
plt.grid()
plt.show()

#Scatter Plot - Battery Storage vs Grid Reliability
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Battery Storage (%)", y="Grid Reliability Index", hue="Load Type", data=df, palette="coolwarm")
plt.title("Battery Storage vs Grid Reliability")
plt.xlabel("Battery Storage (%)")
plt.ylabel("Grid Reliability Index")
plt.grid()
plt.show()

#Histogram - Power Factor Distribution
plt.figure(figsize=(8, 5))
plt.hist(df["Power Factor"], bins=10, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel("Power Factor")
plt.ylabel("Frequency")
plt.title("Distribution of Power Factor")
plt.grid()
plt.show()

#Pie Chart - Load Type Distribution
load_counts = df["Load Type"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(load_counts, labels=load_counts.index, autopct='%1.1f%%', colors=['skyblue', 'orange', 'lightgreen'], startangle=140)
plt.title("Distribution of Load Types")
plt.show()

#Box Plot - Voltage Variations by Load Type
plt.figure(figsize=(8, 5))
sns.boxplot(x="Load Type", y="Voltage (V)", data=df)
plt.title("Voltage Variations by Load Type")
plt.xlabel("Load Type")
plt.ylabel("Voltage (V)")
plt.grid()
plt.show()

#Heatmap - Correlation Between Parameters
import numpy as np
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Smart Grid Parameters")
plt.show()

#Fault Detection Analysis - Bar Chart
plt.figure(figsize=(6, 4))
df["Fault Detection"].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title("Fault Detection Status")
plt.xlabel("Fault Detected")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid()
plt.show()
```
